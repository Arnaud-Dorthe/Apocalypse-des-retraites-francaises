# Notes de bugs

* Dans les graphiques (par exemple sur https://blogs.mediapart.fr/bruno-scherrer/blog/201219/dapres-le-cor-philippe-vous-donne-la-liberte-de-partir-62-ans-et-detre-pauvre), le titre "Niveau de vie par rapport à celui des actifs" doit être remplacé par "Niveau de vie par rapport à l'ensemble de la population". On le voit rapidement, car c'est la courbe qui commence approximativement à 1 et qui descend vers 0.75. 
* Dans la description du modèle, le bilan financier modifié $S_S$ utilise $T_s$ et pas $T$. En d'autres termes, l'équation :

$$
S_s =  B [T-K(P_s+dP)]
$$

doit être remplacée par :

$$
S_s =  B [T_s-K(P_s+dP)].
$$

* Dans le Pilotage 1 : calcul à âge et niveau de vie défini, l'équation qui évalue $P_s$ doit utiliser $S_s$ et pas $S$. 
En d'autres termes, l'équation :

\begin{align}
P_s = \frac{U-S/B-KdP}{Z+K}
\end{align}

doit être remplacée par :

\begin{align}
P_s = \frac{U-S_s/B-KdP}{Z+K}.
\end{align}


* Dans le Pilotage 1 : calcul à âge et niveau de vie défini, l'équation qui définit $Z$ doit utiliser $R_{NVs}$ et pas $R_{NV}$. L'équation :

$$
Z = \frac{ (1-T_{PR}) C_{NV} }{R_{NV}}
$$

doit être remplacée par :

$$
Z = \frac{ (1-T_{PR}) C_{NV} }{R_{NVs}}
$$

* Dans le pilotage 2 : calcul à cotisations et pensions définies, l'équation définissant $K$ doit utiliser $S_s$ et non pas $S$. En d'autres termes, l'équation :

$$
K = \frac{T_s-S/B}{P_s+dP}.
$$

doit être remplacée par :

$$
K = \frac{T_s-S_s/B}{P_s+dP}.
$$

* Dans le pilotage 4 : calcul à cotisations et âge définis, l'équation :

$$
P_s = \frac{T-S_s/B}{K}-dP
$$

doit être remplacée par :

$$
P_s = \frac{T_s-S_s/B}{K}-dP.
$$

* Dans tous les calculs de pilotage, c'est $S_s$ qu'on utilise en entrée, et pas $S$.
